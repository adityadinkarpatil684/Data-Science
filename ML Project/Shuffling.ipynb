```python
import pandas as pd

# Load dataset
df = pd.read_csv("loan_preprocess_dataset.csv")

# 1. Check distribution of target column in order
print(df['loan_status'].head(20))   # look at first 20
print(df['loan_status'].tail(20))   # look at last 20

# 2. Plot sequence of target values
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.plot(df['loan_status'].values, marker='o', linestyle='none')
plt.title("Sequence of Loan Status (row-wise)")
plt.xlabel("Row index")
plt.ylabel("Loan Status")
plt.show()
```

```python
# Shuffle the dataset
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Save shuffled dataset to new CSV
df_shuffled.to_csv("loan_preprocess_dataset_shuffled.csv", index=False)

print("✅ Shuffled dataset saved as 'loan_preprocess_dataset_shuffled.csv'")
```
